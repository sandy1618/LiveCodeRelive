```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple example signal (number of cars passing by)
n_samples = 100
time = np.arange(n_samples)
signal = np.sin(0.1 * time) + np.random.normal(scale=0.1, size=n_samples)

# Calculate auto-correlation
max_lag = 20
auto_corr = np.correlate(signal, signal, mode='full') / n_samples

# Plot the original signal and auto-correlation
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(time, signal)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Signal Amplitude')

plt.subplot(2, 1, 2)
lags = np.arange(-len(auto_corr)//2, len(auto_corr)//2 )
plt.plot(lags, auto_corr)
plt.title('Auto-correlation')
plt.xlabel('Time Lag')
plt.ylabel('Auto-correlation Value')

plt.tight_layout()
plt.show()
```
